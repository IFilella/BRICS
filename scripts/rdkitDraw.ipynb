```python
import mol
import rdkit.Chem as Chem
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem.Draw.MolDrawing import MolDrawing, DrawingOptions
from rdkit.Chem import Draw
```

```python
DrawingOptions.bondLineWidth=1.8
```

```python
smile = 'Cc1cc(-c2csc(N=C(N)N)n2)cn1C'
inchi = 'InChI=1S/C10H13N5S/c1-6-3-7(4-15(6)2)8-5-16-10(13-8)14-9(11)12/h3-5H,1-2H3,(H4,11,12,13,14)'
```

## SMILE

```python
m = mol.mol(smile=smile)
m.mol
```

```python
m.get_BRICSdecomposition()
m.get_clean_fragments()
frags = [Chem.MolFromSmiles(frag) for frag in m.cfragments]
Draw.MolsToGridImage(frags)
```

## InChI

```python
m = mol.mol(InChI=inchi)
m.mol
```

```python
m.get_BRICSdecomposition()
m.get_clean_fragments()
frags = [Chem.MolFromSmiles(frag) for frag in m.cfragments]
Draw.MolsToGridImage(frags)
```
